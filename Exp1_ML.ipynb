```python
import tensorflow as tf
g =tf.Graph()
x=2
y=2
with g.as_default():

  op1=tf.add(x,y)
  op2=tf.multiply(x,y)
  op3=tf.pow(op2,op1)
print(op3)

with tf.compat.v1.Session(graph=g) as sess:
  result=sess.run(op3)
print(op3)
```
