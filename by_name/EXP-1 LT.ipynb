```python
import tensorflow as tf

g = tf.Graph()

with g.as_default():
       a = tf.constant([1.0], name="a")
       b = tf.constant([2.0], name="b")
       c = tf.add(a, b, name="addition")

print(c)
with tf.compat.v1.Session(graph=g) as sess:
    result = sess.run(c)
print(result)
```

```python
import tensorflow as tf

g = tf.Graph()
x = 2
y = 2

with g.as_default():
  op1 = tf.add(x, y)
  op2 = tf.multiply(x, y)
  op3 = tf.pow(op2, op1)
  print(op3)

  with tf.compat.v1.Session(graph = g) as sess:
    result = sess.run(op3)
    print(op3)
    print(result)
```
